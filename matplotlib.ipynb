### 주석과 그림 추가

```python
from datetime import datetime

fig = plt.figure(figsize=(5, 4))
ax = fig.add_subplot(1, 1, 1)

data = pd.read_csv('ch08/spx.csv', index_col=0, parse_dates=True)
spx = data['SPX']

spx.plot(ax=ax, style='k-')
```

주석과 글자는 text, arrow, annotate 함수를 사용해 추가할 수 있다. 아래는 text 함수의 사용 예이다.
* ax.text(x, y, 'Hello world!', family='monospace', fontsize=10)

```python
crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75),
                xytext=(date, spx.asof(date) + 225),
                arrowprops=dict(facecolor='red', headwidth=4, width=2, headlength=4),
                horizontalalignment='left', verticalalignment='top')
```

```python
# Zoom in on 2007-2010
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])
ax.set_title('Important dates in the 2008-2009 financial crisis')
```

### 막대 그래프

수직 막대 그래프를 그리려면 kind='bar', 수평 막대 그래프를 그리려면 kind='barh'로 지정한다. 이 경우 Series나 DataFrame의 색인은 X(bar) 눈금이나 Y(barh) 눈금으로 사용된다.

```python
plt.close('all')
%matplotlib inline
```

```python
fig, axes = plt.subplots(2, 1, figsize=(6, 5))
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))
data.plot(kind='bar', ax=axes[0], color='y', alpha=0.7)
data.plot(kind='barh', ax=axes[1], color='y', alpha=0.7)
```

plot.bar 또는 plot.barh 메소드를 사용해도 된다.

DataFrame에서 막대 그래프는 각 로우의 값을 함께 묶어서 하나의 그룹마다 각각의 막대를 보여준다.

DataFrame의 칼럼 이름인 'Genus'가 범례의 제목으로 사용되었다. 쌓인 막대 그래프는 stacked=True 옵션을 지정하여 생성한다.

### 산포도

산포도는 2개의 1차원 데이터 묶음 간의 관계를 나타내고자 할 때 유용하다. matplotlib의 scatter 메소드를 사용해 산포도를 그린다. 

[Statsmodel 프로젝트] ( http://statsmodels.sourceforge.net/ )에서 macrodata 데이터 셋을 불러온 다음 몇 가지 변수를 선택하고 로그차를 구해보자.

```python
macro = pd.read_csv('ch08/macrodata.csv')
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]
trans_data = np.log(data).diff().dropna()
trans_data[-5:]
```

```python
fig = plt.figure()
```

```python
plt.scatter(trans_data['m1'], trans_data['unemp'])
plt.title('Changes in log %s vs. log %s' % ('m1', 'unemp'))
```

pandas의 scatter_matrix 함수를 이용하면 DataFrame 객체로부터 산포도 행렬을 간단히 생성할 있다. 또한 pandas는 대각선을 따라 각 변수에 대한 히스토그램이나 밀도 그래프도 생성할 수 있다. 

```python
%matplotlib inline
```

```python
plt.close('all')
```

```python
fig = plt.figure()
pd.plotting.scatter_matrix(trans_data, diagonal='kde', color='r', alpha=0.5, figsize=(8, 6))
```

```python
sns.regplot('m1', 'unemp', data=trans_data)
plt.title('Changes in log %s versus log %s' % ('m1', 'unemp'))
```

데이터를 탐색하는 단계에서 같은 그룹안의 변수들에 대하여 산포도를 보면 도움이 된다. 이를 a pairs plot 또는 scatter plot matrix라 하는데 seaborn 모듈은 편리한 pairplot 함수를 갖고 있다.

```python
sns.pairplot(trans_data, diag_kind='kde', plot_kws={'alpha': 0.2})
```

### Facet Grids and Categorical Data

다양한 카테고리의 변수를 가진 데이터를 시각화하는 한 방법은 facet grid를 이용하는 것이다. seaborn은 factorplot이라는 내장함수를 갖고 있는데 이를 이용하면 간단하게 다양한 종류의 faceted plot을 만들 수 있다.

```python
sns.factorplot(x='day', y='tip_pct', hue='time', col='smoker',
               kind='bar', data=tips[tips.tip_pct < 1])
```
